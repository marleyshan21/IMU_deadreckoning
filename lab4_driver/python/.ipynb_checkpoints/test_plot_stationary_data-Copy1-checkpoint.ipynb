```python
import bagpy
from bagpy import bagreader
import pandas as pd
import seaborn as sea
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
def data_csv(b):

    csvfiles = []
    for t in b.topics:
        data = b.message_by_topic(t)
        csvfiles.append(data)

    print(csvfiles[0])
    data = pd.read_csv(csvfiles[0])
    return data
```

```python
df_gps
```

```python
#median data

df_gps['UTM_easting_median'] = df_gps['UTM_easting'] - df_gps['UTM_easting'].median(axis=0)
df_gps['UTM_northing_median'] = df_gps['UTM_northing'] - df_gps['UTM_northing'].median(axis=0)
```

```python
df_gps['UTM_easting'].median(axis=0), df_gps['UTM_northing'].median(axis=0)
```

```python
df_gps['UTM_easting_median'], df_gps['UTM_northing_median'] 
```

```python
cep_radius = 0.59*(df_gps['UTM_easting_median'].std() + df_gps['UTM_northing_median'].std())
cep_radius
```

```python
drms_2 = 2* np.sqrt(df_gps['UTM_easting_median'].var() + df_gps['UTM_northing_median'].var())
drms_2
```

```python
#corrected - the median of the each data in easting and northing was subtracted 
#from the  easting and northing datapoints respectively. 
fig, ax = bagpy.create_fig(1)
ax[0].grid(True)
ax[0].axis('equal')
ax[0].scatter(x='UTM_easting_median', y='UTM_northing_median', data=df_gps)
plt.xlabel("UTM_easting_corrected")
plt.ylabel("UTM_northing_corrected")
cir_CEP = plt.Circle((0,0),cep_radius,color='forestgreen',alpha=0.4)
cir_drms = plt.Circle((0,0),drms_2,color='orange',alpha=0.4)
ax[0].add_artist(cir_CEP)
ax[0].add_artist(cir_drms)
plt.show()
```

```python
from scipy.stats import norm
```

```python
mu, std = norm.fit(df_gps['UTM_easting_median'])
```

```python

df_gps['UTM_easting_mean'] = df_gps['UTM_easting'] - df_gps['UTM_easting'].mean()
df_gps['UTM_easting_mean']
```

```python
mu, std = norm.fit(df_gps['UTM_easting_mean']) 
  
# Plot the histogram.
plt.hist(df_gps['UTM_easting_mean'], bins=10, density=True, alpha=0.6, color='b')
  
# Plot the PDF.
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
  
plt.plot(x, p, 'k', linewidth=2)
title = "Fit Values: {:.2f} and {:.2f}".format(mu, std)
plt.title(title)
  
plt.show()
```

```python
mu, std = norm.fit(df_gps['UTM_easting_median']) 
  
# Plot the histogram.
plt.hist(df_gps['UTM_easting_median'], bins=10, density=True, alpha=0.6, color='b')
  
# Plot the PDF.
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
  
plt.plot(x, p, 'k', linewidth=2)
title = "Fit Values: {:.2f} and {:.2f}".format(mu, std)
plt.title(title)
  
plt.show()
```

```python
#percentage of points near the median
#considering that the median here (0,0)
df_gps['dist_euclidean'] = np.sqrt( (df_gps['UTM_easting_median'])**2 + (df_gps['UTM_northing_median'])**2)

df_gps['dist_euclidean'] 
```

```python
#  Distance Root Mean Squared (DRMS) - 
#DRMS is the square root of the average of the square errors 
df_gps['square_error'] = (df_gps['UTM_easting_median'])**2 + (df_gps['UTM_northing_median'])**2
drms = np.sqrt(df_gps['square_error'].mean())
```
